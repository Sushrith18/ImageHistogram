```python
import cv2
import matplotlib.pyplot as plt

# Read the image (include extension!)
image = cv2.imread("dip-img.jpg")

# Check if loaded properly
if image is None:
    print("Error: Image not loaded. Check file name and extension.")
else:
    print("Image loaded successfully")
    print("Image shape:", image.shape)

    # Convert BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display image
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.title("Loaded Image")
    plt.show()

    # Print top-left 5x5 pixel matrix
    print("Top-left 5x5 pixel matrix:\n")
    print(image[0:5, 0:5])

```
